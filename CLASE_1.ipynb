**L√≥gica Proposicional**

Es una rama fundamental de la l√≥gica que se enfoca en el an√°lisis de la validez de argumentos y proposiciones.

**Definici√≥n de Proposici√≥n**

Una proposici√≥n es una afirmaci√≥n que puede ser verdadera o falsa, pero no ambas.Ejemplo: El sol es una estrella.

**Operadores L√≥gicos**

Son como las piezas maestras que nos permiten combinar proposiciones simples y construir enunciados complejos con significados precisos. Al igual que los bloques de construcci√≥n, estos operadores nos dan la flexibilidad de crear estructuras l√≥gicas s√≥lidas y analizar argumentos con rigor.

**Los operadores l√≥gicos principales son:**

**Negaci√≥n**(¬¨)

**Conjunci√≥n (y) (‚àß)** Se hace verdadera si todos son V

# **Ejemplos en Python con los Operadores L√≥gicos:**

Para empezar a realizra prograci√≥n en python sobre los operadores l√≥gicos debemos tener en cuenta algunas definciones, que son la siguientes:

**1. ‚úè Comparaciones b√°sicas:**
Las comparaciones son la base de la l√≥gica computacional. En Python, podemos utilizar operadores de comparaci√≥n para evaluar la relaci√≥n entre dos valores:

**==: Igualdad** (comprueba si dos valores son exactamente iguales)

**!=: Desigualdad** (comprueba si dos valores no son exactamente iguales)

**">": Mayor que** (comprueba si un valor es mayor que otro)

**<: Menor que** (comprueba si un valor es menor que otro)

**">=:" Mayor o igual que** (comprueba si un valor es mayor o igual que otro)

**<=: Menor o igual que** (comprueba si un valor es menor o igual que otro)

**Ejemplo en Python**


```
edad = 18
es_mayor_de_edad = edad >= 18

print(f"¬øEs mayor de edad? {es_mayor_de_edad}")  # Salida: ¬øEs mayor de edad? True
```

**2. ‚úè Operadores l√≥gicos binarios:**

Los operadores l√≥gicos binarios nos permiten combinar dos valores booleanos para obtener un nuevo valor booleano. Entre los m√°s comunes encontramos:

**and:** Conjunci√≥n (ambos valores deben ser True para que la expresi√≥n sea True)

**or:** Disyunci√≥n (al menos uno de los valores debe ser True para que la expresi√≥n sea True)

**not:** Negaci√≥n (invierte el valor del booleano original)

**Ejemplo en Python**

```
esta_lloviendo = True
hace_sol = False

puede_ir_a_la_playa = not esta_lloviendo and hace_sol

print(f"¬øPuede ir a la playa? {puede_ir_a_la_playa}")  # Salida: ¬øPuede ir a la playa? True
```

**3. ‚úè Operadores l√≥gicos anidados:**

Los operadores l√≥gicos pueden anidarse para crear expresiones a√∫n m√°s complejas. Las reglas de precedencia determinan el orden de evaluaci√≥n:

**not** se eval√∫a primero

**and y or** se eval√∫an de izquierda a derecha

**Ejemplo en Python**

```
esta_lloviendo = True
hace_sol = False
tiene_sombrilla = True

puede_ir_al_parque = (not esta_lloviendo or hace_sol) and tiene_sombrilla

print(f"¬øPuede ir al parque? {puede_ir_al_parque}")  # Salida: ¬øPuede ir al parque? True
```

**4. ‚úè Funciones l√≥gicas:**

Python ofrece funciones l√≥gicas integradas para realizar operaciones m√°s espec√≠ficas:

**any():** Eval√∫a si al menos un elemento en un iterable es True

**all():** Eval√∫a si todos los elementos en un iterable son True

**Ejemplo en Python**

```
frutas = ["manzana", "naranja", "pera"]

hay_manzana = any(fruta == "manzana" for fruta in frutas)
son_todas_frutas = all(fruta in ["manzana", "naranja", "pera"] for fruta in frutas)

print(f"¬øHay manzana en la lista? {hay_manzana}")  # Salida: ¬øHay manzana en la lista? True
print(f"¬øSon todas las frutas manzana, naranja o pera? {son_todas_frutas}")  # Salida: ¬øSon todas las frutas manzana, naranja o pera? True
```

# üî¥ **Cuantificadores**
Son palabras o s√≠mbolos que nos permiten expresar proposiciones que abarcan todos o algunos elementos de un conjunto.

**Tipos de cuantificadores:**

Existen dos tipos principales de cuantificadores:

**Cuantificador universal (‚àÄ):** Representado por "para todo", "todo" o "cada", el cuantificador universal indica que la proposici√≥n es verdadera para todos los elementos del conjunto sin excepci√≥n. Por ejemplo: "‚àÄx ‚àà ‚Ñù, x^2 ‚â• 0" (para todo n√∫mero real x, x al cuadrado es mayor o igual que 0).

 **Cuantificador existencial (‚àÉ)**: Representado por "existe", "hay" o "alguno", el cuantificador existencial indica que la proposici√≥n es verdadera para al menos un elemento del conjunto. Por ejemplo: "‚àÉx ‚àà ‚Ñ§, x^2 = 2" (existe un n√∫mero entero x tal que x al cuadrado es igual a 2).

**Negaci√≥n de proposiciones cuantificadas:**

La negaci√≥n de una proposici√≥n cuantificada se realiza invirtiendo el cuantificador y negando el predicado.

* La negaci√≥n de "‚àÄx ‚àà ‚Ñù, x^2 ‚â• 0" es "‚àÉx ‚àà ‚Ñù, x^2 < 0" (existe un n√∫mero real x tal que x al cuadrado es menor que 0).

* La negaci√≥n de "‚àÉx ‚àà ‚Ñ§, x^2 = 2" es "‚àÄx ‚àà ‚Ñ§, x^2 ‚â† 2" (para todo n√∫mero entero x, x al cuadrado es diferente de 2).

**Ejemplos en Python**

**1. Cuantificador universal (‚àÄ):**

Para simular el cuantificador universal "‚àÄx ‚àà D, P(x)", donde "D" es el conjunto de discurso y "P(x)" es el predicado, podemos utilizar un bucle **FOR** que recorre cada elemento en "D" y verifica si el predicado "P(x)" es verdadero para cada uno. Si el predicado es falso para al menos un elemento, la proposici√≥n completa se considera falsa. En caso contrario, si el predicado es verdadero para todos los elementos, la proposici√≥n se considera verdadera

```
def es_par(x):
  return x % 2 == 0

def todos_pares(conjunto):
  for numero in conjunto:
    if not es_par(numero):
      return False
  return True

conjunto1 = {2, 4, 6, 8}
conjunto2 = {1, 3, 5}

print(todos_pares(conjunto1))  # Salida: True
print(todos_pares(conjunto2))  # Salida: False
```

**2. Cuantificador existencial (‚àÉ):**

Para simular el cuantificador existencial "‚àÉx ‚àà D, P(x)", podemos utilizar un bucle FOR que recorre cada elemento en "D" y verifica si el predicado "P(x)" es verdadero para al menos un elemento. Si encontramos un elemento para el cual el predicado es verdadero, la proposici√≥n completa se considera verdadera. En caso contrario, si el predicado es falso para todos los elementos, la proposici√≥n se considera falsa.

```
def es_primo(x):
  if x <= 1:
    return False
  for i in range(2, int(x**0.5) + 1):
    if x % i == 0:
      return False
  return True

def existe_primo(conjunto):
  for numero in conjunto:
    if es_primo(numero):
      return True
  return False

conjunto1 = {2, 3, 4, 5}
conjunto2 = {1, 6, 8, 9}

print(existe_primo(conjunto1))  # Salida: True
print(existe_primo(conjunto2))  # Salida: False
```

# üîª**Teor√≠a de Conjuntos**
 Es una rama fundamental de las matem√°ticas y la l√≥gica, se encarga de estudiar los conjuntos, agrupaciones abstractas de objetos que comparten alguna caracter√≠stica en com√∫n. A trav√©s de un lenguaje formal y riguroso, esta teor√≠a nos permite analizar las propiedades de los conjuntos, las relaciones entre ellos y las operaciones que podemos realizar sobre ellos.

**Definici√≥n de Conjunto**

Un conjunto es una colecci√≥n de elementos distintos. Los elementos pueden ser n√∫meros, letras, u otros objetos.

**Notaci√≥n**:

Un conjunto se denota con llaves: {a, b, c}

Conjunto vac√≠o: {}

**Operaciones con Conjuntos**

Uni√≥n (A ‚à™ B)

Intersecci√≥n (A ‚à© B)

Diferencia (A - B)

Complemento (A^c)

**Ejemplo en Python**

```
conjunto_A = {1, 2, 3}
conjunto_B = {3, 4, 5}

conjunto_union = conjunto_A | conjunto_B
print(conjunto_union)  # Salida: {1, 2, 3, 4, 5}
```

```
conjunto_A = {1, 2, 3}
conjunto_B = {3, 4, 5}

conjunto_interseccion = conjunto_A & conjunto_B
print(conjunto_interseccion)  # Salida: {3}
```

```
conjunto_A = {1, 2, 3}
conjunto_B = {3, 4, 5}

conjunto_diferencia = conjunto_A - conjunto_B
print(conjunto_diferencia)  # Salida: {1, 2}
```
