## üé≤ Imagen y Preimagen de una Funci√≥n
Cuando hablamos de funciones, es crucial entender los conceptos de imagen y preimagen:

### ‚ôü Preimagen
La preimagen de un conjunto en \(B\) es el conjunto de todos los elementos en \(A\) que se mapean a ese conjunto espec√≠fico en \(B\) bajo la funci√≥n \(f\). La preimagen de un elemento \(y\) en \(B\) se denota como \(f^{-1}(y)\). Este conjunto de valores se denomina Dominio de la funci√≥n.

### üìå Imagen
La imagen de un conjunto \(A\) bajo una funci√≥n \(f\) es el conjunto de todos los valores que \(f(x)\) puede tomar cuando \(x\) var√≠a en \(A\). Matem√°ticamente, se denota como \(f(A)\). Este conjunto de valores se conoce como el Rango o Recorrido de la funci√≥n.

# **üìç Funci√≥n Compuesta**
Una funci√≥n compuesta es una nueva funci√≥n que se obtiene al combinar dos o m√°s funciones existentes. Se forma aplicando una funci√≥n a la salida de otra funci√≥n. La funci√≥n compuesta se representa con el s√≠mbolo ‚àò, donde f ‚àò g denota la funci√≥n compuesta formada por las funciones f y g.

**Notaci√≥n**

La notaci√≥n para una funci√≥n compuesta f ‚àò g significa que primero se aplica la funci√≥n g a la entrada, y luego se aplica la funci√≥n f al resultado de g. Matem√°ticamente, se expresa como:

$( f(x) $) y luego $( g $) a $( f(x) $).

 image.png

**üß© Ejemplo de funci√≥n compuesta en Python:**


```
def funcion_interna(x):
  return x ** 2

def funcion_externa(y):
  return 2 * y + 1

def funcion_compuesta(x):
  return funcion_externa(funcion_interna(x))

# Ejemplo de uso
resultado = funcion_compuesta(3)
print(resultado)  # Salida: 13
```
