# **üìã Sistemas N√∫mericos**

Los sistemas num√©ricos son conjuntos de s√≠mbolos y reglas que utilizamos para representar y manipular cantidades.

Cada sistema num√©rico tiene sus propias caracter√≠sticas, como:

Base: La unidad fundamental a partir de la cual se construyen los dem√°s n√∫meros. Por ejemplo, el sistema decimal tiene una base de 10, mientras que el sistema binario tiene una base de 2.

S√≠mbolos: Los elementos que representan las unidades b√°sicas del sistema. En el sistema decimal, los s√≠mbolos son los d√≠gitos del 0 al 9.

Valor posicional: En algunos sistemas, como el decimal, el valor de un s√≠mbolo depende de la posici√≥n que ocupa dentro del n√∫mero. Por ejemplo, en el n√∫mero 123, el 1 vale 100, el 2 vale 20 y el 3 vale 3.

Reglas de operaci√≥n: Las normas que definen c√≥mo realizar operaciones aritm√©ticas como la suma, resta, multiplicaci√≥n y divisi√≥n.

# **‚ô¶ Ejemplo en Python**
C√°lculos en sistemas num√©ricos:


```python
def sumar_binarios(binario1, binario2):
  """
  Suma dos n√∫meros binarios.

  Args:
    binario1: El primer n√∫mero binario como una cadena de caracteres.
    binario2: El segundo n√∫mero binario como una cadena de caracteres.

  Returns:
    str: La suma de los dos n√∫meros binarios como una cadena de caracteres.
  """
  resultado = ""
  carry = 0
  i = len(binario1) - 1
  j = len(binario2) - 1
  while i >= 0 or j >= 0:
    suma_digitos = 0
    if i >= 0:
      suma_digitos += int(binario1[i])
    if j >= 0:
      suma_digitos += int(binario2[j])
    suma_digitos += carry

    if suma_digitos >= 2:
      digito = suma_digitos % 2
      carry = 1
    else:
      digito = suma_digitos
      carry = 0

    resultado = str(digito) + resultado
    i -= 1
    j -= 1
  if carry:
    resultado = "1" + resultado
  return resultado

numero_binario1 = input("Ingrese el primer n√∫mero binario: ")
numero_binario2 = input("Ingrese el segundo n√∫mero binario: ")
suma = sumar_binarios(numero_binario1, numero_binario2)
print(f"La suma de {numero_binario1} y {numero_binario2} es: {suma}")
```

# **‚ôü Conceptos b√°sicos de Cardinalidad**
Conjunto: Una colecci√≥n de elementos bien definidos y distinguibles entre s√≠. Los elementos de un conjunto pueden ser n√∫meros, objetos, ideas o cualquier otra entidad.

**Elemento:** Cada uno de los componentes que forman parte de un conjunto.

**Cardinalidad:** El n√∫mero de elementos que contiene un conjunto. Se representa con el s√≠mbolo # o con la letra n.

**üïπTipos de cardinalidad:**

**Cardinalidad finita:** Un conjunto finito es aquel que tiene un n√∫mero determinado y contable de elementos. Se puede contar cada elemento del conjunto, uno por uno.

Ejemplo: Un conjunto que contiene las letras del alfabeto ingl√©s tiene una cardinalidad finita de 26 elementos.

**Cardinalidad infinita:** Un conjunto infinito es aquel que tiene un n√∫mero incontable de elementos. No se puede contar cada elemento del conjunto, uno por uno.

Ejemplo: El conjunto de todos los n√∫meros naturales (1, 2, 3, ...) es un conjunto infinito.

**üîé Propiedades de la cardinalidad:**

**Cardinalidad del conjunto vac√≠o:** El conjunto vac√≠o, que no contiene ning√∫n elemento, tiene una cardinalidad de 0.
Cardinalidad de un conjunto unitario: Un conjunto unitario, que contiene un solo elemento, tiene una cardinalidad de 1.

**Cardinalidad de un subconjunto:** La cardinalidad de un subconjunto de un conjunto nunca puede ser mayor que la cardinalidad del conjunto original.

**Cardinalidad de la uni√≥n de dos conjuntos:** La cardinalidad de la uni√≥n de dos conjuntos disjuntos (que no comparten elementos comunes) es igual a la suma de las cardinalidades de los conjuntos individuales.

# **üòé Ejemplo en Python**


```python
# Ejemplo 1: Cardinalidad de una lista usando un bucle for
conjunto_animales = ["perro", "gato", "conejo"]
cardinalidad = 0

for animal in conjunto_animales:
  cardinalidad += 1

print(f"La cardinalidad del conjunto de animales es: {cardinalidad}")  # Salida: 3

# Ejemplo 2: Cardinalidad de un conjunto usando un bucle for
conjunto_letras = {"a", "b", "c", "d"}
cardinalidad = 0

for letra in conjunto_letras:
  cardinalidad += 1

print(f"La cardinalidad del conjunto de letras es: {cardinalidad}")  # Salida: 4
```

# **Producto Cratesiano üìä**
El producto cartesiano de dos conjuntos A y B, representado como A x B, se define como el conjunto que contiene todos los pares ordenados posibles, donde el primer elemento pertenece a A y el segundo elemento pertenece a B.
El producto cartesiano se puede representar de diversas maneras:

**Notaci√≥n por pares ordenados**: {(a, b) | a ‚àà A, b ‚àà B}

Notaci√≥n matricial: [Elemento 1 de A, Elemento 2 de A, ..., Elemento n de A] x [Elemento 1 de B, Elemento 2 de B, ..., Elemento m de B]

**‚öñ Propiedades del producto cartesiano:**

**Propiedad conmutativa:** El orden de los conjuntos en el producto cartesiano no altera el resultado. Es decir, A x B = B x A.

**Propiedad asociativa:** El producto cartesiano de m√°s de dos conjuntos se puede asociar de cualquier manera sin afectar el resultado. Es decir, (A x B) x C = A x (B x C).

# **‚≠ï Ejemplos del Producto Cartesiano en Python**


```python
import itertools

# Ejemplo 1: Producto cartesiano de dos listas
conjunto_colores = ["rojo", "verde", "azul"]
conjunto_formas = ["cuadrado", "c√≠rculo", "tri√°ngulo"]

producto_cartesiano = list(itertools.product(conjunto_colores, conjunto_formas))
print(producto_cartesiano)  # Salida: [('rojo', 'cuadrado'), ('rojo', 'c√≠rculo'), ('rojo', 'tri√°ngulo'), ('verde', 'cuadrado'), ..., ('azul', 'tri√°ngulo')]

# Ejemplo 2: Producto cartesiano de un conjunto y un rango
conjunto_letras = "abc"
rango_numeros = range(1, 4)

producto_cartesiano = list(itertools.product(conjunto_letras, rango_numeros))
print(producto_cartesiano)  # Salida: [('a', 1), ('a', 2), ('a', 3), ('b', 1), ..., ('c', 3)]
```
