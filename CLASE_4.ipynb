# **Tipos de Funciones**
- *Inyectiva (uno a uno):* Una funci贸n $(f: A \to B$) es inyectiva si $(f(a_1) = f(a_2)$) implica $(a_1 = a_2$).
- *Sobreyectiva (sobre):* Una funci贸n $(f: A \to B$) es sobreyectiva si para cada $(b \in B$) existe al menos un $(a \in A$) tal que $(f(a) = b$).
- *Biyectiva:* Una funci贸n es biyectiva si es inyectiva y sobreyectiva.

# ** Ejemplo en Python**


```python
def verificar_funcion(funcion, dominio, codominio):
  """
  Funci贸n que verifica si una funci贸n dada es inyectiva, sobreyectiva o biyectiva.

  Args:
    funcion: La funci贸n a evaluar (una funci贸n que recibe un elemento del dominio y devuelve su imagen en el codominio).
    dominio: El conjunto del cual la funci贸n toma sus elementos de entrada.
    codominio: El conjunto donde la funci贸n env铆a sus elementos de salida.

  Returns:
    Un diccionario con las siguientes claves:
      "inyectiva": True si la funci贸n es inyectiva, False en caso contrario.
      "sobreyectiva": True si la funci贸n es sobreyectiva, False en caso contrario.
      "biyectiva": True si la funci贸n es biyectiva, False en caso contrario.
  """

  # Conjunto para almacenar las im谩genes ya evaluadas
  imagenes_evaluadas = set()

  # Bandera para determinar inyectividad
  es_inyectiva = True

  # Recorremos cada elemento del dominio
  for elemento_dominio in dominio:
    # Calculamos la imagen del elemento actual
    imagen = funcion(elemento_dominio)

    # Verificamos si la imagen ya ha sido evaluada
    if imagen in imagenes_evaluadas:
      # Si la imagen ya ha sido evaluada, la funci贸n no es inyectiva
      es_inyectiva = False
      break

    # Agregamos la imagen al conjunto de im谩genes evaluadas
    imagenes_evaluadas.add(imagen)

  # Bandera para determinar sobreyectividad
  es_sobreyectiva = True

  # Recorremos cada elemento del codominio
  for elemento_codominio in codominio:
    # Verificamos si existe alg煤n elemento del dominio que tenga como imagen el elemento actual del codominio
    existe_preimagen = False
    for elemento_dominio in dominio:
      if funcion(elemento_dominio) == elemento_codominio:
        existe_preimagen = True
        break

    # Si no existe preimagen para el elemento actual del codominio, la funci贸n no es sobreyectiva
    if not existe_preimagen:
      es_sobreyectiva = False
      break

  # Determinamos si la funci贸n es biyectiva
  es_biyectiva = es_inyectiva and es_sobreyectiva

  # Devolvemos un diccionario con los resultados
  return {
      "inyectiva": es_inyectiva,
      "sobreyectiva": es_sobreyectiva,
      "biyectiva": es_biyectiva
  }

# Ejemplo de uso
funcion_ejemplo = lambda x: x**2  # Funci贸n a evaluar
dominio_ejemplo = {1, 2, 3}  # Dominio de la funci贸n
codominio_ejemplo = {1, 4, 9}  # Codominio de la funci贸n

resultado = verificar_funcion(funcion_ejemplo, dominio_ejemplo, codominio_ejemplo)

print(f"Funci贸n inyectiva: {resultado['inyectiva']}")
print(f"Funci贸n sobreyectiva: {resultado['sobreyectiva']}")
print(f"Funci贸n biyectiva: {resultado['biyectiva']}")
```
