# **🔻Funciones**
**Conceptos Básicos:**

+ Una función se define como una relación entre dos conjuntos: el dominio (conjunto de valores de entrada válidos) y el codominio (conjunto de valores de salida posibles).

+ A cada valor de entrada del dominio le corresponde un único valor de salida en el codominio.

+ Las funciones se representan matemáticamente utilizando una notación que incluye el nombre de la función, los argumentos y el valor de salida.

image.png

# **📕 Definición**
Una función $(f$) de un conjunto $(A$) a un conjunto $(B$) es una relación que asigna a cada elemento de $(A$) exactamente un elemento de $(B$). Se denota como $(f: A \to B$).

**Ejemplo**

La función $(f(x) = x^2$) asigna a cada número real $(x$) su cuadrado.

# **🎲Tipos de Funciones**
- *Inyectiva (uno a uno):* Una función $(f: A \to B$) es inyectiva si $(f(a_1) = f(a_2)$) implica $(a_1 = a_2$).
- *Sobreyectiva (sobre):* Una función $(f: A \to B$) es sobreyectiva si para cada $(b \in B$) existe al menos un $(a \in A$) tal que $(f(a) = b$).
- *Biyectiva:* Una función es biyectiva si es inyectiva y sobreyectiva.

# **📝 Ejemplo en Python**


```python
def verificar_funcion(funcion, dominio, codominio):
  """
  Función que verifica si una función dada es inyectiva, sobreyectiva o biyectiva.

  Args:
    funcion: La función a evaluar (una función que recibe un elemento del dominio y devuelve su imagen en el codominio).
    dominio: El conjunto del cual la función toma sus elementos de entrada.
    codominio: El conjunto donde la función envía sus elementos de salida.

  Returns:
    Un diccionario con las siguientes claves:
      "inyectiva": True si la función es inyectiva, False en caso contrario.
      "sobreyectiva": True si la función es sobreyectiva, False en caso contrario.
      "biyectiva": True si la función es biyectiva, False en caso contrario.
  """

  # Conjunto para almacenar las imágenes ya evaluadas
  imagenes_evaluadas = set()

  # Bandera para determinar inyectividad
  es_inyectiva = True

  # Recorremos cada elemento del dominio
  for elemento_dominio in dominio:
    # Calculamos la imagen del elemento actual
    imagen = funcion(elemento_dominio)

    # Verificamos si la imagen ya ha sido evaluada
    if imagen in imagenes_evaluadas:
      # Si la imagen ya ha sido evaluada, la función no es inyectiva
      es_inyectiva = False
      break

    # Agregamos la imagen al conjunto de imágenes evaluadas
    imagenes_evaluadas.add(imagen)

  # Bandera para determinar sobreyectividad
  es_sobreyectiva = True

  # Recorremos cada elemento del codominio
  for elemento_codominio in codominio:
    # Verificamos si existe algún elemento del dominio que tenga como imagen el elemento actual del codominio
    existe_preimagen = False
    for elemento_dominio in dominio:
      if funcion(elemento_dominio) == elemento_codominio:
        existe_preimagen = True
        break

    # Si no existe preimagen para el elemento actual del codominio, la función no es sobreyectiva
    if not existe_preimagen:
      es_sobreyectiva = False
      break

  # Determinamos si la función es biyectiva
  es_biyectiva = es_inyectiva and es_sobreyectiva

  # Devolvemos un diccionario con los resultados
  return {
      "inyectiva": es_inyectiva,
      "sobreyectiva": es_sobreyectiva,
      "biyectiva": es_biyectiva
  }

# Ejemplo de uso
funcion_ejemplo = lambda x: x**2  # Función a evaluar
dominio_ejemplo = {1, 2, 3}  # Dominio de la función
codominio_ejemplo = {1, 4, 9}  # Codominio de la función

resultado = verificar_funcion(funcion_ejemplo, dominio_ejemplo, codominio_ejemplo)

print(f"Función inyectiva: {resultado['inyectiva']}")
print(f"Función sobreyectiva: {resultado['sobreyectiva']}")
print(f"Función biyectiva: {resultado['biyectiva']}")
```

# **🔴 Tipos de Demostración**
**🧩 Demostración Directa**
Una demostración directa es aquella en la que se parte de las hipótesis y mediante un razonamiento lógico se llega a la conclusión deseada.
Claro, aquí tienes un ejemplo de demostración directa de una propiedad de una función matemática.

**Ejemplo**

**Proposición:**

La función \( f(x) = 2x + 3 \) es estrictamente creciente.

**Definición de estrictamente creciente:**

Una función \( f(x) \) es estrictamente creciente si para cualquier \( x_1 \) y \( x_2 \) en su dominio, si \( x_1 < x_2 \) entonces \( f(x_1) < f(x_2) \).

**Demostración directa:**

1. Sea \( f(x) = 2x + 3 \).

2. Supongamos que \( x_1 \) y \( x_2 \) son dos números en el dominio de \( f \) y que \( x_1 < x_2 \).

3. Evaluamos la función en \( x_1 \) y \( x_2 \):
   \[
   f(x_1) = 2x_1 + 3
   \]
   \[
   f(x_2) = 2x_2 + 3
   \]

4. Queremos mostrar que \( f(x_1) < f(x_2) \). Consideramos la diferencia:
   \[
   f(x_2) - f(x_1) = (2x_2 + 3) - (2x_1 + 3)
   \]

5. Simplificando la expresión:
   \[
   f(x_2) - f(x_1) = 2x_2 + 3 - 2x_1 - 3 = 2x_2 - 2x_1 = 2(x_2 - x_1)
   \]

6. Dado que \( x_2 > x_1 \), \( x_2 - x_1 \) es positivo. Multiplicando por 2 (un número positivo) no cambia el signo, por lo tanto:
   \[
   2(x_2 - x_1) > 0
   \]

7. Entonces, \( f(x_2) - f(x_1) > 0 \), lo que implica que \( f(x_1) < f(x_2) \).

**Conclusión:**

La función \( f(x) = 2x + 3 \) es estrictamente creciente porque hemos demostrado directamente que para cualquier \( x_1 < x_2 \), \( f(x_1) < f(x_2) \).

# **✨Preimagen e Imagen**

**⭕ Imagen:** Conjunto de números reales que son imagen por de los elementos de su dominio.

**⭕ Preimagen:** Es un conjunto de puntos en  A  cuya imagen pertenece a ese conjunto.

Supongamos que tenemos la función:

f:A ->  B , x->  f(x)

Sea  E⊆A ,  f(E) :{ f(x)ϵB:xϵE }, significa que  f(E)  es un conjunto de las imagenes cuyos elemetos de  E  existen en  B , es decir la imagen de E.
Sea  F⊆B ,  f−1(F):= { xϵA:f(x)ϵF }, esto seria la preimagen de F (osea que podemos usar la función inversa para encontrar la preimagen).

**Ejemplos:**

$f(A)⊆B$

$f(R)$={$f(x)ϵR:xϵR$}, = {$x^2+1:xϵR$}

$f(R)$={$xϵR: x≥1$} o $[1,+∞[⊆R$

La preimagen seria: $f^-1(]-10,0[)$ y la definición seria: {$xϵR:x^2+1ϵ]-10,0[$}
