# **üéâEjemplo en PyThon**

```python
def ordenamiento_burbuja(lista):
  n = len(lista)
  for i in range(n):
    for j in range(0, n - i - 1):
      if lista[j] > lista[j + 1]:
        lista[j], lista[j + 1] = lista[j + 1], lista[j]

lista_desordenada = [5, 2, 4, 1, 3]
print(f"Lista desordenada: {lista_desordenada}")
ordenamiento_burbuja(lista_desordenada)
print(f"Lista ordenada: {lista_desordenada}")
```

# **ü§† ¬øC√≥mo Localizar R√°pidamente un N√∫mero en una Lista Extensa?**
Se lo puede haer medinate las siguientes b√∫squedas:

**B√∫squeda Lineal**

‚Ä¢ La b√∫squeda lineal implica revisar cada
elemento secuencialmente hasta encontrar el
objetivo. Este m√©todo es simple pero puede
ser ineficiente en listas grandes.

‚Ä¢ El n√∫mero m√°ximo de pasos para encontrar un
n√∫mero ser√≠a n = 100.

 **Ejmeplo en Python**


```python
def buscar_lineal(lista, elemento):
  """
  Busca un elemento dentro de una lista utilizando b√∫squeda lineal.

  Args:
    lista: Lista de elementos a buscar.
    elemento: Elemento a buscar dentro de la lista.

  Returns:
    bool: True si el elemento se encuentra en la lista, False en caso contrario.
  """
  for i in range(len(lista)):
    if lista[i] == elemento:
      return True
  return False

# Ejemplo de uso
lista_nombres = ["Juan", "Mar√≠a", "Pedro", "Ana", "Carlos"]
elemento_a_buscar = "Ana"

resultado_busqueda = buscar_lineal(lista_nombres, elemento_a_buscar)

if resultado_busqueda:
  print(f"El elemento {elemento_a_buscar} SI se encuentra en la lista.")
else:
  print(f"El elemento {elemento_a_buscar} NO se encuentra en la lista.")
```

**B√∫squeda Binaria**

‚Ä¢ La b√∫squeda binaria divide el conjunto de
datos a la mitad en cada paso, lo que reduce
significativamente el n√∫mero de
comparaciones necesarias, especialmente en
conjuntos grandes.

‚Ä¢ El n√∫mero m√°ximo de pasos es
aproximadamente log2 n . En este caso ser√≠a:
log2 100 = 6.64 que se aproxima a 7.

**Ejemplo en Python**

```python
def busqueda_binaria(lista, elemento):
  """
  Busca un elemento en una lista ordenada utilizando b√∫squeda binaria.

  Args:
    lista: Lista ordenada de elementos.
    elemento: Elemento a buscar en la lista.

  Returns:
    int: √çndice del elemento en la lista si se encuentra, -1 si no se encuentra.
  """
  izquierda = 0
  derecha = len(lista) - 1

  while izquierda <= derecha:
    mitad = (izquierda + derecha) // 2
    dato_mitad = lista[mitad]

    if dato_mitad == elemento:
      return mitad
    elif dato_mitad < elemento:
      izquierda = mitad + 1
    else:
      derecha = mitad - 1

  return -1

# Ejemplo de uso
lista_ordenada = [3, 8, 15, 22, 37, 49, 56, 62]
elemento_a_buscar = 37

indice = busqueda_binaria(lista_ordenada, elemento_a_buscar)

if indice != -1:
  print(f"El elemento {elemento_a_buscar} se encuentra en la posici√≥n {indice} de la lista.")
else:
  print(f"El elemento {elemento_a_buscar} NO se encuentra en la lista.")
```

**Convertir de decimal a binario:** Se divide el n√∫mero decimal por 2 repetidamente, anotando el residuo en cada paso. Los residuos, en orden inverso, representan el n√∫mero binario.

**Convertir de binario a decimal:** Se multiplica cada d√≠gito binario por su correspondiente potencia de 2 (2^0, 2^1, etc.) y se suman los resultados.

**Conversi√≥n de binario a decimal:**

Este ejemplo convierte un n√∫mero binario a su representaci√≥n decimal utilizando el **m√≥dulo math y la funci√≥n int():**

```python
def convertir_binario_a_decimal(numero_binario):
  """
  Convierte un n√∫mero binario a su representaci√≥n decimal.

  Args:
    numero_binario: El n√∫mero binario a convertir (como cadena).

  Returns:
    int: El valor decimal del n√∫mero binario.
  """
  return int(numero_binario, 2)

numero_binario = input("Ingrese un n√∫mero binario: ")
resultado_decimal = convertir_binario_a_decimal(numero_binario)
print(f"El n√∫mero binario {numero_binario} en decimal es: {resultado_decimal}")
```
